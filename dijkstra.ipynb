```
import heapq
def dijkstra(edges, num_v, goal):
    dist = [float('inf')] * num_v
    dist[0] = 0
    q = []
    heapq.heappush(q, [0, [0]])

    while len(q) > 0:
        # ヒープから取り出し
        a, u = heapq.heappop(q)
        last = u[-1]
        print('a:',a)
        print('q:',q)
        print('u:',u)
        print('last:',last)

        if last == goal:
            return u
        for i in edges[last]:
            print('edges[last]:',edges[last])
            if dist[i[0]] > dist[last] + i[1]:
                # 頂点までのコストが更新できれば更新してヒープに登録
                dist[i[0]] = dist[last] + i[1]
                heapq.heappush(q, [dist[i[0]], u + [i[0]]])
                print('dist[i[0]]:',dist[i[0]])
                print('[dist[i[0]], u + [i[0]]]:',[dist[i[0]], u + [i[0]]])
                print('=========================')

    return q
```

```
edges = [
          [[1,8]],
          [[2,2],[3,17]],
          [[3,15],[4,7]],
          [[5,1]],
          [],
          [[6,3],[7,5]],
          [[8,6],[9,12]],
          [[16,16],[10,20]],
          [],
          [[11,4],[12,1]],
          [[13,5],[14,2]],
          [],
          [[15,3],[16,3]],
          [],
          [[17,2],[18,2]],
          [],
          [[19,8]],
          [],
          [[20,2],[21,8]],
          [],
          [],
          []
]
```

```
#@title デフォルトのタイトル テキスト
from time import time
start_time = time()
dijkstra(edges, 22, 19)
proc_time = time() - start_time
print("実行時間は：", proc_time,"秒")
```
